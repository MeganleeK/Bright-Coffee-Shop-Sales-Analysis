### 5. Calculating Revenue



```python
df.columns
```

### 5.1. Calculating the Total Revenue

```python
total_revenue=df['TOTAL_AMOUNT'].sum()
print(total_revenue)
```

### 5.2. Calculating Revenue by Product

```python
revenue_by_product=df.groupby('PRODUCT_DETAIL')['TOTAL_AMOUNT'].sum().sort_values(ascending=False)
print(revenue_by_product)
```

### 5.3.Calculating Revenue by Location

```python
revenue_by_location=df.groupby('STORE_LOCATION')['TOTAL_AMOUNT'].sum().sort_values(ascending=False)
print(revenue_by_location)
```

### 5.4. Calculating Revenue by Time of Day

```python
revenue_by_time=df.groupby('TIME_BUCKETS')['TOTAL_AMOUNT'].sum().sort_values(ascending=False)
print(revenue_by_time)
```

### 6.Data Visualizations

### 6.1. Top 10 products by Revenue

```python
top_products = df.groupby('PRODUCT_DETAIL')['TOTAL_AMOUNT'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(
    x=top_products.values,
    y=top_products.index,
    palette="YlOrBr" )
plt.title('Top 10 Revenue-Generating Products')
plt.xlabel('Total Revenue')
plt.ylabel('Product')
plt.show()
```

### 6.2. Product Category Performance

```python
cat_sales = df.groupby('PRODUCT_CATEGORY')['TOTAL_AMOUNT'].sum()

fig = px.pie(values=cat_sales.values, names=cat_sales.index, title='Sales by Product Category')
fig.show()
```

### 6.3.Store Location Comparison

```python
location_sales = df.groupby('STORE_LOCATION')['TOTAL_AMOUNT'].sum().sort_values()
plt.figure(figsize=(10,6))
sns.barplot(x=location_sales.values, y=location_sales.index, palette='magma')
plt.title("Revenue by Store Location")
plt.xlabel("Total Revenue")
plt.ylabel("Location")
plt.show()
```

### 6.4 Revenue by Time of Day

```python
time_revenue = df.groupby('TIME_BUCKETS')['TOTAL_AMOUNT'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=time_revenue.index, y=time_revenue.values, palette='muted')
plt.title("Revenue by Time of Day")
plt.xlabel("Time Bucket")
plt.ylabel("Total Revenue")
plt.show()
```
